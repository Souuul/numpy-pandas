```python
# 이렇게 Database나 CSV파일로부터 데이터를 읽어들여서 DataFrame으로 만들었어요!!
# 이렇게 만든 DataFrame을 JSON으로 저장할 수 있어요!!
# JSON 파일로 저장할 때 4가지 방식이 있어요!!
# Database안의 book table에서 원하는 내용의 데이터를 가져온 후 DataFrame으로 만들어서 이 DataFrame을 JSON 파일로 저장해 보아요!

import pymysql.cursors
import pandas as pd

conn = pymysql.connect(host= 'localhost',
                      user = 'data',
                      password = 'data',
                      db='library',
                      charset='utf8')


sql = 'SELECT btitle,bauthor, bprice FROM book WHERE btitle like "%여행%"'

df = pd.read_sql(sql, con = conn)

display(df)

# DataFrame을 json파일로 저장할 때 총 4가지 서로 다른 형식이 존재.
# Unicode로 파일을 생성한 후 데이터를 저장해야 한글이 정상 처리가 되요!
# 파일을 저장할 폴더는 미리 생성이 되어 있어야 합니다.!!

# with : 일반적으로 파일 처리할 때 순서는 파일열기 - 내용쓰기 - 파일닫기 이순서...
#        with 구문을 이용하면 resource의 close처리(해제처리)가 자동으로 이루어져요!
with open('./data/json/books_columns.json', 'w', encoding='utf8') as file:
    df.to_json(file, force_ascii=False, orient = 'columns')
    # orient = columns로 설정하면 df의 column이 json의 key값으로 들어가요.
    
with open('./data/json/books_records.json', 'w', encoding='utf8') as file:
    df.to_json(file, force_ascii=False, orient = 'records')
    
with open('./data/json/books_index.json', 'w', encoding='utf8') as file:
    df.to_json(file, force_ascii=False, orient = 'index')

with open('./data/json/books_values.json', 'w', encoding='utf8') as file:
    df.to_json(file, force_ascii=False, orient = 'values')
```

```python
# 지금부터는 row indexing에 대해서 알아보아요!
import numpy as np
import pandas as pd
data = {'이름' : ['이지은', '박동훈', '홍길동', '강감찬', '오혜영'],
       '학과': ['컴퓨터', '기계', '철학', '컴퓨터', '철학'],
       '학년' : [1,2,2,4,3],
       '학점' : [1.5, 2.0, 3.1, 1.1, 2.7]}

df = pd.DataFrame(data, columns = ['학과', '이름', '학점', '학년', '등급' ],
                 index=['one', 'two', 'three', 'four', 'five'])
display(df)

# column indexing
# 단일 column을 indexing
# print(df['이름']) #OK. Series로 리턴
# print(df['학점':'학과']) # error column은 slicing을 지원하지 않아요!
# display(df[['학과','이름']]) # OK column에 대해서 Fancy indexing은 가능

# row indexing - part.1 행에 대한 숫자 index를 이용하는 방법
# print(df['one']) # Error // 'one'이란 이름의 column을 찾는 것 
# print(df[0]) # Error // .row에 대해서 숫자 index로 단일 indexing 은 안됩니다.
# display(df[1:3]) # row 는 slicing이 가능!!!! - slicing한 결과는 view. DataFrame으로 리턴
# display(df[1:])   # OK 
# display(df[[0,2]]) # Error. row 에 대해서 index 숫자를 이용한 Fancy indexing 은 안되요!

# row indexing - part.2 행에 대한 별도의 index를 이용하는 방법
# df['one'] # Error.
display(df['one' : 'three']) #index를 이용한 row slicing 가능
display(df['three' : ]) #index를 이용한 row slicing 가능
# display(df['three' : -1 ]) #Error. 숫자 index와 index를 혼용해서 슬라이싱 할수 없어요
# display(df['two', 'four']) # Error. Fancy indexing 안되요!!

# row indexing - part.3 (loc[]를 row indexing. 가장 많이 사용하는 방법 중 하나.)
# loc를 이용할때는 숫자 index가 아닌 부여한 index를 이용
# 지정한 index만 가능합니다!! 숫자 index는 사용할 수 없음 
print(df.loc['one']) # OK !!. loc와 index를 이용하면 단일 row를 추출할 수 있어요.
                     # Series로 return

display(df.loc['one' : 'three']) # slicing 도 가능
display(df.loc[['one' , 'three']]) # Fancy indexing 도 가능


# row indexing - part.4 (iloc[]를 row indexing. 가장 많이 사용하는 방법 중 하나.)
# loc vs iloc .. iloc[] 무조건 숫자 index만 사용할 수 있어

print(df.iloc[1]) # OK !!. loc와 index를 이용하면 단일 row를 추출할 수 있어요.

display(df.loc['one' : 'three','이름':'학년']) # 열도 추출가능
display(df.loc['one' : 'three',['이름', '학년']]) # 열의 Fancy indexing 도 가능
# 학점이 1.5을 초과하는 학생의 이름과 학점을 DataFrame으로 출력하세요!!

scale = df[df['학점']>1.5]

display(scale[['이름', '학점']])

display(df.loc[df['학점']>1.5,['이름', '학점']])
```

```python
## Series와 DataFrame 의 indexing 방법!!
# DataFrame이 제공하는 함수들!! (집계함수, 통계기반 함수..)

# 기댓값 (expected value) : 어떤 확률을 가진 사건을 무한히 반복했을 경우 얻을 수 있는 값의 평균으로 기대할 수 있는 값
# 주사위 1개를 던지는 사건을 무한히 반복했을 떄 기댓값은 얼마인가요?

import numpy as np
result = np.random.randint(1,7,(1000000,))
print(result.mean())

# 편차 (deviation) : 확률변수 x와 평균(기댓값)의 차이!!
# 국민 개개인의 소득.. 홍길동, 김길동, 아이유, 김연아, 놀부, ...
# 국민 평균소득을 산출
# 홍길동소득 - 평균소득, 김길동소득 - 평균소득, .... 놀부소득 - 평균소득
# 편차의 관점에서 보면 양극화가 심하다!!
# 편차를 이용하면 데이터의 흩어진 정도를 알기 쉬워요!!
# 편차의 가장 큰 단점..
# 데이터의 흩어진 정도를 하나의 숫자로 표현하기 힘들어요!!
# 편차의 합은 0
# 분산(variance) : 데이터의 흩어진 정도를 알기위해서 사용하는 편차의 제곱의 평균!
# 편차의 제곱의 평균! 제곱을 이용했기 때문에 사용하기가 애매

# std(standard deviation ): 표준편차!! => 분산의 제곱근.

# import numpy as np
# arr = np.array([4,6,1,3,8,8], dtype=np.int32)
# print (arr)
# print(arr.sum())
# print(arr.mean())
# print(arr.var())
# print(arr.std())


####

# 공분산 (covariance)
# 두 개의 확률변수의 관계를 보여줄 때 사용하는 값.
# 공분산은 두 확률변수의 편차의 곱에 대한 평균
# 확률변수 X(독립변수)와 Y(종속변수)에 대해 X가 변할떄 Y가 변하는 정도.

#그래프를 이용해서 공분산의 의미를 파악해보아요!!
import numpy as np
import matplotlib.pyplot as plt

# 독립변수 X에 대해 종속변수 Y의 값을 랜덤으로 생성해서 알아보아요!
# np.random.seed(1)

# x = np.random.randint(-20, 20,(10,))
# y = np.random.randint(-10, 10,(10,))

# x_mean = x.mean() # X의 평균
# y_mean = y.mean() # Y의 평균

# # 이렇게 구한 값을 가지고.. scatter(산점도)
# plt.scatter(x,y,color='red')
# plt.scatter(x.mean(),y.mean(), color='blue')
# plt.show()

# 공분산은 데이터의 변화량에 대한 총합을 의미
# 공분산이 양수인지, 음수인지에 따라서 데이터의 방향성을 알 수 있어요 !
# 공분산의 단점 > 단위의 문제때문에 연관성은 알수 있는데 그 강도는 알 수 없어요!

# 만약 확률변수 X와 Y 가 독립이면 공분산을 계산하면 0에 수렴
# 역은 성립하지 않아요! 공분산이 0이면 두 확률변수는 독립이에요 라고 할 수 없어요!

# 간단하게 공분산을 계산해 보아요!
np.random.seed(2)
sampleNum = 100 # 데이터의 개수

x = np.random.randint(0, 10,(sampleNum,))
y = np.random.randint(-20, 20,(sampleNum,))

x_mean = x.mean() # X의 평균
y_mean = y.mean() # Y의 평균

# 공분산은 편차의 곱의 평균
x_deviation = x - x_mean
y_deviation = y - y_mean

result = 0

for tmp in range(sampleNum):
    result += x_deviation[tmp] * y_deviation[tmp]
    
result_covariance = result / (sampleNum -1) # 표준공분산.
print(result_covariance)
print(np.cov(x,y)) # numpy 로 제공되는 것을 사용가능
```

```python
# DataFrame의 결합(merge) => Database에 Table join 기능
# DataFrame 두개를 연결시켜서 새로운 DataFrame을 만들어보아요!!
import numpy as np
import pandas as pd


# merge : inner join 어떤기준으로 결합이 되는지 
# df1, df2를 inner join으로 결합합니다. 학번을 기준으로 결합합니다. (교집합) 

data1 = {'학번':[1,2,3,4],
        '이름' : ['이지안', '박동훈', '이순신', '강감찬'],
        '학년' : [2,4,1,3]}

data2 = {'학번':[1,2,4,5],
        '학과': ['CS', 'MATH','MATH','CS'],
        '학점' : [3.4,2.9,4.5,1.2]}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
display(df1)
display(df2)
df3 = pd.merge(df1, df2) # 공통적인 부분만 추출해서 새로운 DataFrame을 생성 (자동)
df3 = pd.merge(df1, df2, on='학번', how='inner') # 학번을 기준으로 inner join
display(df3)


# merge : outer join 합집합을 구함
# full outer join > Union 합집합
# left outer join > 차집합! (df1 - df2) + intersect // 차집합과 교집합의 합 ()
# right outer joint > 차집합 df2 - df1
df3 = pd.merge(df1, df2, on='학번', how='outer') # 공통적인 부분만 추출해서 새로운 DataFrame을 생성 (자동)
display(df3) # 없는것은 NaN
df3 = pd.merge(df1, df2, on='학번', how='left') # left outerjoin
display(df3) # 없는것은 NaN
```

```python
# 이렇게 merge에 대한 기본내용을 이해했으면
# 나머지는 여러가지 경우에 대한 사용법
# Column 명이 다를경우
import numpy as np
import pandas as pd
data1 = {'학번':[1,2,3,4],
        '이름' : ['이지안', '박동훈', '이순신', '강감찬'],
        '학년' : [2,4,1,3]}

data2 = {'학생학번':[1,2,4,5],
        '학과': ['CS', 'MATH','MATH','CS'],
        '학점' : [3.4,2.9,4.5,1.2]}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
display(df1)
display(df2)
df3 = pd.merge(df1, df2, left_on='학번',right_on='학생학번', how='inner') # 학번을 기준으로 inner join
display(df3)
```

```python
# Series를 1차원, 2차원으로 연결하는 방법에 대해서 알아보아요!!
# concatenation(연결)

import numpy as np
import pandas as pd 
s1 = pd.Series([0,1], index = ['a', 'c'])
s2 = pd.Series([4,3,2], index = ['b','c','e'])
s3 = pd.Series([4,6], index=['f','g'])

# Series는 1차원 vector형태에요!
# 연결하는 방법은 당연히 2가지 방법이 있을꺼에요!!
# 행방향으로 연결하는 방법, 열방향으로 연결하는 방법.

print(pd.concat([s1,s2,s3], axis = 0)) # Series를 1차원으로 연결
print(pd.concat([s1,s2,s3], axis = 1, sort= False)) # Series를 1차원으로 연결
#                                        index를 조심해야 해요!!
#                                        sort : index를 정렬할 것인지 아닌지..

print(pd.concat([s1,s2,s3], axis = 1, sort= True)) # Series를 1차원으로 연결
```

```python
# 중복행 처리

import numpy as np
import pandas as pd

df = pd.DataFrame({'k1' : ['one'] *3 + ['two']*4,
                  'k2' : [1,1,2,3,3,4,4]})

display(df)
# DataFrame 에서 중복항에 대한 boolean mask를 추출
print(df.duplicated()) # 처음나오면 중복이여도 False로 나옴
'''
0    False
1     True
2    False
3    False
4     True
5    False
6     True
'''
# 중복행을 추출해보아요!
df.loc[df.duplicated(), :]

display(df.drop_duplicates()) # 중복행을 찾아서 지우고 그 결과를 DataFrame을 리턴


df = pd.DataFrame({'k1' : ['one'] *3 + ['two']*4,
                  'k2' : [1,1,2,3,3,4,4],
                  'k3' : np.arange(7)})
display(df)
display(df.drop_duplicates()) # 중복행이 없어요 ( 전부다 같아야 중복행)
display(df.drop_duplicates(['k1'])) # 'k1' 기준으로 중복되는 것을 전부삭제
display(df.drop_duplicates(['k1', 'k2'])) # 'k1' 과 'k2'기준으로 중복되는 것을 전부삭제
```

```python
### Pandas 마지막 내용이에요!!!
### DataFrame Grouping 에 대하여 알아보아요!!

import numpy as np
import pandas as pd
my_dict ={'학과':['컴퓨터', '체육교육과', '컴퓨터', '체육교육과', '컴퓨터'],
         '학년' : [1,2,3,2,3],
         '이름' : ['홍길동', '김연아','최길동', '아이유', '신사임당'],
         '학점' : [1.5,4.4,3.7,4.5,3.8]}
df = pd.DataFrame(my_dict)
display(df)

# 학과를 기준으로 Grouping 작업을 할건데... 학점을 grouping
# Series를 Grouping 하는 개녕
dept = df['학점'].groupby(df['학과'])
display(dept) # pandas.core.groupby.generic.SeriesGrouBy object

# 그룹안에 데이터를 확인하고 싶은 경우 get_group() 사용
dept.get_group('컴퓨터') #Series
 
display(dept.get_group('컴퓨터')) #Series

# 각 그룹의 사이즈를 확인하고 싶은 경우 size()
print (dept.size()) #Series로 리턴, 각 그룹의 count
print(dept.mean()) # Series로 리턴

# 2단계 grouping 도 가능해요!!

dept_year = df['학점'].groupby([df['학과'], df['학년']])
display(dept_year.mean())

# unstack() : 최하위 index를 column으로 설정!!
display(dept_year.mean().unstack())
```

```python
# Grouping을 하면 왠지.. 해당 그룹을 반복해서 어떠한 처리를 하고 싶어요!!
import numpy as np
import pandas as pd
my_dict ={'학과':['컴퓨터', '체육교육과', '컴퓨터', '체육교육과', '컴퓨터'],
         '학년' : [1,2,3,2,3],
         '이름' : ['홍길동', '김연아','최길동', '아이유', '신사임당'],
         '학점' : [1.5,4.4,3.7,4.5,3.8]}
df = pd.DataFrame(my_dict)
display(df)

# 학과로 Grouping을 한 후 for 문을 이용해서 반복처리

for ((dept,year), group) in df.groupby(['학과', '학년']):
    print(dept)
    print(year)
    display(group)
    print('===============')
```
