```python
# Pandas Exercise - 1

import numpy as np
import pandas as pd

df_movies = pd.read_csv('./data/movies.csv')

df_rate = pd.read_csv('./data/ratings.csv')

df1 = pd.DataFrame(df_movies)
df2 = pd.DataFrame(df_rate)
df3 = pd.merge(df1, df2, on='movieId', how='inner')
display(df3)

# 1. 사용자가 평가한 모든 영화의 전체 평균 평점
print(1)
print(df3['rating'].mean()) #3.501556983616962

# 2. 각 사용자별 평균 평점

df_group_dept = df2.groupby(df2['userId'])
print(2)
display(df_group_dept['rating'].mean())

# 3. 각 영화별 평균 평점
df_group_movieid = df3.groupby('title', as_index=True)
print(3)
df_group_movieid = df3.groupby(df3['title'])

df_mean_movie = df_group_movieid.mean().sort_values(by = 'movieId' ,ascending= True)
display(df_mean_movie[['movieId','rating']])




# 4. 평균 평점이 가장 높은 영화의 제목(동률이 있을 경우 모두 출력)
df_group_movieid = df3.groupby(df3['title'])
df_ever_movie = df_group_movieid[['rating']].mean()
df_max = df_ever_movie.max()


dd = df_ever_movie[df_ever_movie[['rating']]==df_max]

df_na = dd.dropna(how ='any' ,inplace = False) # how= 'any' > NaN이 하나라도 행에 포함되어있으면 행자체를 삭제
print(4)
display(df_na)


# 5. Comedy영화 중 가장 평점이 낮은 영화의 제목
# display(df3)
comedy_df =  df3[df3['genres'].str.contains('Comedy')]
print(5)
# display(comedy_df_min.sort_values(by = 'title' ,ascending= True))
# display(comedy_df[comedy_df['rating'] == comedy_df['rating'].min()]['title'])

comedy_df = comedy_df.groupby(['title'], as_index=False).mean()
display(comedy_df[comedy_df['rating']==comedy_df['rating'].min()])


# 6. 2015년도에 평가된 모든 Romance 영화의 평균 평점은?

df_romance =  df3[df3['genres'].str.contains('Romance')]

df_romance_2015 = df_romance[ df_romance['timestamp'] >= 1420070400]
df_romance_2015 = df_romance_2015[df_romance_2015['timestamp'] < 1451606400]

# display(df_romance_2015)
print(6)
print(df_romance_2015['rating'].mean())
```

```python
# Pandas Exercise - 2

import numpy as np
import pandas as pd

mpg_txt = pd.read_csv('./data/mpg.txt')


df_mpg = pd.DataFrame(mpg_txt)
display(df_mpg)


# 주요컬럼 설명
# manufacturer : 제조회사
# displ : 배기량
# cyl : 실린더 개수
# drv : 구동 방식
# hwy : 고속도로 연비
# class : 자동차 종류
# model : 자동차 모델명
# year : 생산연도
# trans : 변속기 종류
# cty : 도시 연비
# fl : 연료 종류


# 1. displ(배기량)이 4 이하인 자동차와 5 이상인 자동차 중 
# 어떤 자동차의 hwy(고속도로 연비)가 평균적으로 더 높은지 확인하세요.
# under_4 = df_mpg[df_mpg['displ']<=4]['hwy'].mean()
# upper_5 = df_mpg[df_mpg['displ']>=5]['hwy'].mean()
# if under_4 > upper_5:
#     print('배기량이 4이하인 자동차의 평균 고속도로 연비가 높음')
# elif under_4 == upper_5:
#     print('평균 연비가 같음')
# else:
#     print('배기량이 5이상인 자동차가 평균 고속도로 연비가 높음')

    
# 2. 자동차 제조 회사에 따라 도시 연비가 다른지 알아보려고 한다. 
# "audi"와 "toyota" 중 어느 manufacturer(제조회사)의 cty(도시 연비)가 
# 평균적으로 더 높은지 확인하세요.

# audi_cty = df_mpg[df_mpg['manufacturer']=='audi']['cty'].mean()
# toyota_cty = df_mpg[df_mpg['manufacturer']=='toyota']['cty'].mean()

# if audi_cty > toyota_cty:
#     print('audi의 평균 도시연비가 높음')
# elif audi_cty < toyota_cty:
#     print( 'toyota의 평균 도시연비가 높음')
# else:
#     print('toyota와 auid의 평균 도시연비가 같음')
    
    
# 3. "chevrolet", "ford", "honda" 자동차의 고속도로 연비 평균을 알아보려고 한다. 
# 이 회사들의 데이터를 추출한 후 hwy(고속도로 연비) 평균을 구하세요.
chevrolet_cty = df_mpg[df_mpg['manufacturer']=='chevrolet']['hwy'].mean()
ford_cty = df_mpg[df_mpg['manufacturer']=='ford']['hwy'].mean()
honda_cty = df_mpg[df_mpg['manufacturer']=='honda']['hwy'].mean()
print (' chevrolet : {} \n ford : {} \n honda : {}'.format(chevrolet_cty,ford_cty,honda_cty))


# 4. "audi"에서 생산한 자동차 중에 어떤 자동차 모델의 hwy(고속도로 연비)가 
# 높은지 알아보려고 한다. "audi"에서 생산한 자동차 중 hwy가 1~5위에 해당하는 
# 자동차의 데이터를 출력하세요.
audi_df = df_mpg[df_mpg['manufacturer']=='audi'].sort_values(by = 'hwy', ascending= False)
display(audi_df.head())


# 5. mpg 데이터는 연비를 나타내는 변수가 2개입니다. 
# 두 변수를 각각 활용하는 대신 하나의 통합 연비 변수를 만들어 사용하려 합니다. 
# 평균 연비 변수는 두 연비(고속도로와 도시)의 평균을 이용합니다. 
# 회사별로 "suv" 자동차의 평균 연비를 구한후 내림차순으로 정렬한 후 1~5위까지 데이터를 출력하세요.
df_mpg['ever_y'] = df_mpg['hwy'] - df_mpg['cty']
display(df_mpg.sort_values(by = 'ever_y' ,ascending= False).head())



# 6. mpg 데이터의 class는 "suv", "compact" 등 자동차의 특징에 따라 
# 일곱 종류로 분류한 변수입니다. 어떤 차종의 도시 연비가 높은지 비교하려 합니다. 
# class별 cty 평균을 구하고 cty 평균이 높은 순으로 정렬해 출력하세요.
df_class = df_mpg['cty'].groupby(df_mpg['class']).mean()
display(df_class.sort_values(ascending= False))


# 7. 어떤 회사 자동차의 hwy(고속도로 연비)가 가장 높은지 알아보려 합니다. 
# hwy(고속도로 연비) 평균이 가장 높은 회사 세 곳을 출력하세요.
df_manu = df_mpg['hwy'].groupby(df_mpg['manufacturer']).mean()
display(df_manu.sort_values(ascending= False).head(3))




# 8. 어떤 회사에서 "compact" 차종을 가장 많이 생산하는지 알아보려고 합니다. 
# 각 회사별 "compact" 차종 수를 내림차순으로 정렬해 출력하세요.
df_class = df_mpg[df_mpg['class'] == 'compact'].groupby(df_mpg['manufacturer'])
display(df_class.size().sort_values(ascending= False))
```
